Untuk mengubah teks menjadi huruf kecil kita bisa gunakan **lower**, berikut contohnya

```python
kalimat = "Saya sedang belajar Bahasa Pemrograman Python"

lowercase = kalimat.lower()

print(lowercase)
```

Menghapus angka

```python
import re #modul untuk regular expression

kalimat = "Berikut adalah 3 skuter matic favorit indonesia diantaranya Nmax, Vario, dan Beat"

hapus_angka = re.sub(r"\d+", "", kalimat)

print(hapus_angka)
```

Menghapus tanda baca

```python
import string

kalimat = "Saya sedang belajar Bahasa Pemrograman Python!!!"

hapus_tanda_baca = kalimat.translate(str.maketrans("", "", string.punctuation))

print(hapus_tanda_baca)
```

Kode untuk mencoba membaca teks dalam file pdf dan membaca lembar file pdf

```python
import PyPDF2

def baca_pdf(nama_file):

    # buka file pdf dalam mode read binary
    with open(nama_file, "rb") as file:

      # membuat objek pembaca pdf
      pembaca_pdf = PyPDF2.PdfReader(file)

      # mengambil jumlah halaman
      jumlah_halaman = len(pembaca_pdf.pages)
      print(f"Jumlah halaman dalam file PDF: {jumlah_halaman}")

      # membaca setiap halaman
      for halaman in range(jumlah_halaman):
        halaman_pdf = pembaca_pdf.pages[halaman]
        teks = halaman_pdf.extract_text()
        print(f"Teks di halaman {halaman + 1}:\n{teks}\n")

# Contoh penggunaan
baca_pdf("Tugas 1 Digital Marketing.pdf")
```

Berikut adalah kode untuk membaca teks pada file PDF yang sudah dilakukan perubahannya yaitu mengubah semua huruf menjadi huruf kecil dan menghilangkan angka dan tanda baca. Lalu hasilnya akan disimpan dalam format .txt dan ditampilkan dalam command line

```python
import re
import string
import PyPDF2

# Fungsi untuk membaca file PDF dan melakukan pemrosesan teks
def baca_pdf(nama_file):
    # Membuka file PDF dalam mode read-binary (rb)
    with open(nama_file, "rb") as file:
        # Membuat objek pembaca PDF
        pembaca_pdf = PyPDF2.PdfReader(file)

        # Mengambil jumlah halaman di file PDF
        jumlah_halaman = len(pembaca_pdf.pages)
        print(f"Jumlah halaman dalam file PDF: {jumlah_halaman}")

        teks_keseluruhan = ""

        # Membaca setiap halaman PDF
        for halaman in range(jumlah_halaman):
            halaman_pdf = pembaca_pdf.pages[halaman]
            # Ekstrak teks dari halaman
            teks = halaman_pdf.extract_text()
            teks_keseluruhan += teks

        # Case Folding: mengubah teks menjadi huruf kecil
        teks_keseluruhan = teks_keseluruhan.lower()

        # Menghapus angka menggunakan regex
        teks_keseluruhan = re.sub(r"\d+", "", teks_keseluruhan)

        # Menghapus tanda baca
        teks_keseluruhan = teks_keseluruhan.translate(str.maketrans("", "", string.punctuation))

        # Output hasil pemrosesan
        print("Hasil Case Folding, Hapus Angka, dan Hapus Tanda Baca:\n", teks_keseluruhan)

        # Menyimpan hasil pemrosesan ke file teks
        with open("hasil_case_folding.txt", "w") as output_file:
            output_file.write(teks_keseluruhan)

        print("\n Hasil telah disimpan ke dalam file hasil_case_folding.txt")


# Contoh penggunaan fungsi baca_pdf
baca_pdf("Tugas 1 Digital Marketing.pdf")
```
